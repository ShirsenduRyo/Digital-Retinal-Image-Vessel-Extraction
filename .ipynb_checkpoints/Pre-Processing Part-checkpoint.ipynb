```python
def load_data(path):
    """ X = Images and Y = masks """

    train_x = sorted(glob(os.path.join(path, "training", "images", "*.tif")))
    train_y = sorted(glob(os.path.join(path, "training", "1st_manual", "*.gif")))
    return (train_x, train_y)
```

```python
def create_dir(path):
    if not os.path.exists(path):
        os.makedirs(path)
```

```python
def adjust_gamma(image, gamma=1.0):

    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
    for i in np.arange(0, 256)]).astype("uint8")
    return cv2.LUT(image, table)
```

```python
def adjust_square(image, gamma=1.0):
    table = np.array([[i**2 for i in row] for row in image]).astype("uint16")
    return table
```

```python
g = 0
y = 0
```

```python
def augment_thres(images, masks):

    count = 1
    for idx, (x, y) in tqdm(enumerate(zip(images[:1], masks[:1])), total=1):
        """ Extracting names """
        name = x.split("\\")[-1].split(".")[0]
        

        """ Reading image and mask """
        global g
        global y
        src = cv2.imread(x, cv2.IMREAD_COLOR)
        y = imageio.mimread(y)[0]
        
        x = src[:,:,1]
        x = adjust_gamma(x, gamma = 0.6)
#         print(x)
#         print(np.amax(x))
#         x = cv2.GaussianBlur(x, (5, 5), 0)
        
        g = adjust_square(x, gamma=0.6)
#         print("**************")
#         print(g)
#         print(np.amax(g))
#         th_val = threshCal(g)
#         print(th_val)
        #y = imageio.mimread(y)[0]
        
        
        thresh = threshold_otsu(g)
        print("Thresh:",thresh)
        thresh = 4700
        binary = g > thresh

        fig, axes = plt.subplots(ncols=3, figsize=(25, 25))
        ax = axes.ravel()
        ax[0] = plt.subplot(1, 3, 1)
        ax[1] = plt.subplot(1, 3, 2)
        ax[2] = plt.subplot(1, 3, 3, sharex=ax[0], sharey=ax[0])

        ax[0].imshow(x, cmap=plt.cm.gray)
        ax[0].set_title('Original')
        ax[0].axis('off')

        ax[1].imshow(y, cmap=plt.cm.gray)
        ax[1].set_title('Original')
        ax[1].axis('off')
       

        ax[2].imshow(binary, cmap=plt.cm.gray)
        ax[2].set_title('Thresholded')
        ax[2].axis('off')

        plt.show()
        
```
